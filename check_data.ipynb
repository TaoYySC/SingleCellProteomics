```python
tissues = [
    "Bone marrow", "Brain", "Breast", "Cervical", "Colon",
    "Embryo", "Kidney", "Lung", "Ovary", "Pancreas",
    "Peripheral blood", "Pleural effusion", "Skin", "Spinal cord",
    "Testis", "Vascular epithelium"
]

SCP_info = []

for tissue in tissues:
    scp_path = os.path.join('MS-based-SCP',tissue,f"{tissue}.txt")
    with open (scp_path,'r') as file:
        print(f"tissue:{tissue}")
        for line in file:
            line = line.strip()
            print(line)
            scp_info_path = os.path.join('MS-based-SCP',tissue,f"{line}","cell_info.csv")
            print(f"SCP id:{line}")
            sc_info = pd.read_csv(scp_info_path)
            sc_info['SCP ID'] = line
            print(sc_info)
            SCP_info.append(sc_info)      
        
```

```python
SCP_info = pd.concat(SCP_info, ignore_index=True)
SCP_info
```
