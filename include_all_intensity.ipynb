```python
import re
import glob
import pandas as pd
```

```python
def process_raw_data(data):
    data['Protein IDs'] = data['Protein IDs'].str.split(';').str[0]
    data = data[~data['Protein IDs'].str.contains('REV|CON')]
    return data
```

```python
index_map = pd.read_csv('map_result/index_map.csv',index_col = 0)
index_map.drop_duplicates(subset = 'Protein IDs' , inplace = True)
index_map
```

# Bone marrow

## Macrophage

```python
data1 = pd.read_csv('MS-based-SCP/Bone marrow/SCP46620/all_intensity_included_proteinGroups_Primary_BMDM_mpop.txt.csv',index_col = 0)
data2 = pd.read_csv('MS-based-SCP/Bone marrow/SCP46620/all_intensity_included_proteinGroups_Primary_BMDM.txt.csv',index_col = 0)
data1 = process_raw_data(data1)
data2 = process_raw_data(data2)
data1 = data1.merge(
    index_map[['Protein IDs', 'Processed Protein IDs']], 
    on='Protein IDs', 
    how='left' 
)
data1.drop('Protein IDs', axis = 1, inplace = True)
data1.set_index('Processed Protein IDs', inplace = True)
data2 = data2.merge(
    index_map[['Protein IDs', 'Processed Protein IDs']], 
    on='Protein IDs',  
    how='left' 
)
data2.drop('Protein IDs', axis = 1, inplace = True)
data2.set_index('Processed Protein IDs', inplace = True)
data1.columns = data1.columns + '_SCP46620' + '_BMDMmpop' + '_Bone_marrow' + '_Mouse' + '_Macrophage'
data2.columns = data2.columns + '_SCP46620' + '_BMDM' + '_Bone_marrow' + '_Mouse' + '_Macrophage'
merged_data = pd.concat([data1, data2], axis=0, join='outer')
final_data = merged_data.groupby(merged_data.index).first()
final_data = final_data[final_data.index.str.endswith('MOUSE')]
final_data.to_csv('MS-based-SCP/A_Checked_data/BoneMarrow_Mouse_Macrophage.csv')
final_data
```

## Dopaminergic neurons

```python
file_pattern = 'MS-based-SCP/Brain/SCP43381/*all_intensity_included_proteinGroups*.csv'
file_list = glob.glob(file_pattern)
all_data = []
for file in file_list:
    dates = re.search(r'(\d{4}_\d{2}_\d{2}_SC\d{2})', file).group(0)
    data1 = pd.read_csv(file, index_col=0)
    columns = ['Protein IDs','Intensity']
    data1 = data1[columns]
    data1 = process_raw_data(data1)
    data1['Processed Protein IDs'] = data1['Protein IDs'].str.replace(r'^sp\|', '', regex=True)
    data1.drop('Protein IDs', axis=1, inplace=True)
    data1 = data1[[col for col in data1.columns if 'LFQ' not in col]]
    data1.set_index('Processed Protein IDs', inplace=True)
    data1.columns = data1.columns + '_' + dates
    all_data.append(data1)
final_data = pd.concat(all_data, axis=1)
final_data.columns = final_data.columns + '_SCP43381' + '_Brain' + '_Mouse' + '_DopaminergicNeurons'
final_data.to_csv('MS-based-SCP/A_Checked_data/Brain_Mouse_DopaminergicNeurons.csv')
final_data
```
